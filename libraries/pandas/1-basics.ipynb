# Pandas
is a python library to manipulate 2d data (like excel and sheets), called `DataFrame`

It can create a document from:
- lists, dictionaries, etc.
- reading csv
- reading excel

And it can export to them also

```python
import pandas as pd
```

## Creating Document

#### 1. From Lists, Dictionaries, etc.

```python
products = []
products.append({'id': 1, 'title': 'Product A', 'price': 1450})
products.append({'id': 2, 'title': 'Product B', 'price': 2450})
products.append({'id': 3, 'title': 'Product C', 'price': 1350})
products.append({'id': 4, 'title': 'Product D', 'price': 1460})
products.append({'id': 5, 'title': 'Product E', 'price': 5457})
```

```python
my_df = pd.DataFrame(data=products)
```

```python
my_df
```

```python
dict_products = {
  'id': [1, 2, 3, 4, 5, 6],
  'title': ['A', 'B', 'C', 'D', 'E', 'F'],
  'price': [10, 20, 30, 40, 50, 60],
}

another_df = pd.DataFrame(data=dict_products)
another_df
```
