```python
# libs
import numpy as np
import matplotlib.pyplot as plt
```

## Sigmoid

```python
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

## Graph

```python
x = np.linspace(-10, 10, 400)
y = sigmoid(x)

plt.plot(x, y, label='Sigmoid Function')
plt.grid(True)
plt.show()
```

```python
def show_sigmoid_value(target, round_off=2):
  sigmoid_value = sigmoid(target)

  # Plotting the sigmoid function
  plt.plot(x, y, label='Sigmoid Function')
  plt.scatter(target, sigmoid_value, color='red', zorder=5, label=f'Target: {target}, Sigmoid={round(sigmoid_value, round_off)}')

  # Adding labels and title
  plt.xlabel('x')
  plt.ylabel('Sigmoid(x)')
  plt.title('Sigmoid Function with a Specific Value')
  plt.legend()
  plt.grid(True)

  # Showing the plot
  plt.show()
```

```python
show_sigmoid_value(2)
```

```python
show_sigmoid_value(10, 5)
```

```python
show_sigmoid_value(20, 10)
```
