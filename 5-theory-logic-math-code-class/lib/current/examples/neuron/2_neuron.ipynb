### Detailed Use

```python
from src.activation.Sigmoid import Sigmoid
from src.losses.MeanSquaredError import MeanSquaredError
from src.optimizers.StochasticGradientDescent import StochasticGradientDescent
```

```python
neuron = Neuron(
  input_size=X.shape[1],
  learning_rate=0.1,
  activation=Sigmoid(),
  loss_function=MeanSquaredError(),
  optimizer=StochasticGradientDescent(),
  initial_weights=[0.1, 0.2, 0.3, 0.5, 0.4],
  initial_bias=0.5
)
```

```python
neuron.train(
  inputs=X,
  outputs=Y,
  epochs=500,
  verbose=1
)
```

```python
neuron.test(
  inputs=X,
  outputs=Y,
  verbose=0
)
```
