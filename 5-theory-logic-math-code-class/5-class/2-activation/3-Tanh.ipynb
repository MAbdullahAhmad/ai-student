```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Tanh():

  def apply(self, x):
    return np.tanh(x)
  
  def apply_derivative(self, x):
    return 1 - self.apply(x) ** 2
```

```python
tanh = Tanh()
```

```python
x = np.linspace(-10, 10, 100)
y = [tanh.apply(_) for _ in x]

plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.grid()
plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 100)
y = [tanh.apply_derivative(_) for _ in x]

plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.grid()
plt.plot(x, y)
```
