```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Sigmoid():

  def apply(self, x):
    return 1 / (1 + np.exp(-x))
  
  def apply_derivative(self, x):
    sig = self.apply(x)
    return sig * (1 - sig)
```

```python
sigmoid = Sigmoid()
```

```python
x = np.linspace(-10, 10, 100)
y = [sigmoid.apply(_) for _ in x]

plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.grid()
plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 100)
y = [sigmoid.apply_derivative(_) for _ in x]


plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.grid()
plt.plot(x, y)
```
