```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class ReLu():

  def apply(self, x):
    return max(0, x)
  
  def apply_derivative(self, x):
    return 1 if x > 0 else 0
```

```python
relu = ReLu()
```

```python
x = np.linspace(-10, 10, 100)
y = [relu.apply(_) for _ in x]

plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.grid()
plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 100)
y = [relu.apply_derivative(_) for _ in x]

plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.grid()
plt.plot(x, y)
```
