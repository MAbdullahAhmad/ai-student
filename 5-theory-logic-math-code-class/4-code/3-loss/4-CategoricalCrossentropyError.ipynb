```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class CategoricalCrossentropyError():
  def __init__(self, epilon=1e-15):
    self.epilon = epilon

  def calculate(self, y_true, y_pred):
    y_pred = np.clip(y_pred, self.epilon, 1-self.epilon)
    return -np.sum(y_true * np.log(y_pred)) / y_true.shape[0]
```

```python
cce = CategoricalCrossentropyError()
```

```python
x = [0, 0.05, 0.1, 0.5, 0.75, 0.95, 1]
y = [
  cce.calculate(np.array([0, 1, 0]), np.array([0, 1, 0])),
  cce.calculate(np.array([1, 1, 0]), np.array([0.95, 1, 0])),
  cce.calculate(np.array([0, 1, 1]), np.array([0.05, 0.95, 1])),
  cce.calculate(np.array([0, 1, 1]), np.array([1, 0.5, 1])),
  cce.calculate(np.array([0, 1, 1]), np.array([0.75, 0.25, 0.25])),
  cce.calculate(np.array([1, 0, 0]), np.array([0.05, 0.90, 0])),
  cce.calculate(np.array([1, 0, 1]), np.array([0, 1, 0])),
]

plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.grid()
plt.plot(x, y)
```
