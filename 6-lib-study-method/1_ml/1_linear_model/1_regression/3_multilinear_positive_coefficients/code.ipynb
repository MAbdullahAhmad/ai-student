```python
import sys; sys.path.append('../' * 4)
```

```python
from sklearn.linear_model import LinearRegression

from lib.plot.draw_plane      import draw_plane
from lib.plot.draw_3d_scatter import draw_3d_scatter
from lib.plot.draw_line       import draw_line
from lib.plot.draw_scatter    import draw_scatter
from lib.plot.sequential      import sequential
```

```python
from examples.dataset.color_height_weight import dataset
```

```python
dataset.head()
```

```python
X = dataset[dataset.columns[:2]].values
X1, X2, Y = dataset[dataset.columns[0]].values, dataset[dataset.columns[1]].values, dataset[dataset.columns[2]].values
```

```python
draw_3d_scatter(None, X1.tolist(), X2.tolist(), Y.tolist())
```

## Multilinear Regression

```python
regressor = LinearRegression()
```

```python
regressor.fit(X, Y)
```

```python
draw_plane(None, regressor.coef_, regressor.intercept_, elev=20)
```

```python
plt = draw_3d_scatter(None, X1.tolist(), X2.tolist(), Y.tolist())
draw_plane(plt, regressor.coef_, regressor.intercept_, range=[(min(X1), max(X1)), (min(X2), max(X2))], elev=17, azim=70)
```

```python
regressor.coef_, regressor.intercept_
```

```python
draw_line(slope=float(regressor.coef_[1]), intercept=float(regressor.intercept_))
```

```python
sequential(None, [
  ['draw_scatter', {'x':X2.tolist(), 'y':Y.tolist()}],
  ['draw_line', {'slope':float(regressor.coef_[1]), 'intercept':float(regressor.intercept_), 'range': [0, max(X2.tolist() + Y.tolist())]}],
])
```

## Positive Coefficients Model

```python
regressor = LinearRegression(positive=True)
```

```python
regressor.fit(X, Y)
```

```python
draw_plane(None, regressor.coef_, regressor.intercept_, elev=20)
```

```python
plt = draw_3d_scatter(None, X1.tolist(), X2.tolist(), Y.tolist())
draw_plane(plt, regressor.coef_, regressor.intercept_, range=[(min(X1), max(X1)), (min(X2), max(X2))], elev=17, azim=70)
```

```python
regressor.coef_, regressor.intercept_
```

```python
draw_scatter(x=X2.tolist(), y=Y.tolist())
```

```python
draw_line(slope=float(regressor.coef_[1]), intercept=float(regressor.intercept_))
```

```python
sequential(None, [
  ['draw_scatter', {'x':X2.tolist(), 'y':Y.tolist()}],
  ['draw_line', {'slope':float(regressor.coef_[1]), 'intercept':float(regressor.intercept_), 'range': [0, max(X2.tolist() + Y.tolist())]}],
])
```
