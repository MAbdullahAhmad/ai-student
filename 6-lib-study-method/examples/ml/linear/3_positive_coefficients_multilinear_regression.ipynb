## Imports

```python
from lib.ml.linear.regressor.MultilinearRegressor import MultilinearRegressor as LR
from lib.ml.linear.regressor.PositiveCoefficientsMultilinearRegressor import PositiveCoefficientsMultilinearRegressor as PLR
from lib.plot.draw_plane      import draw_plane
from lib.plot.draw_3d_scatter import draw_3d_scatter
from lib.plot.draw_line       import draw_line
from lib.plot.draw_scatter    import draw_scatter
from lib.plot.sequential      import sequential
```

## Dataset

```python
from examples.dataset.color_height_weight import dataset
```

```python
dataset.head()
```

```python
X = dataset[dataset.columns[:2]].values
X1, X2, Y = dataset[dataset.columns[0]].values, dataset[dataset.columns[1]].values, dataset[dataset.columns[2]].values
```

```python
draw_3d_scatter(None, X1.tolist(), X2.tolist(), Y.tolist())
```

## MultilinearRegression Model

```python
regressor = LR()
```

```python
regressor.train(X, Y)
```

```python
draw_plane(None, regressor.coefficients, regressor.intercept, elev=20)
```

```python
plt = draw_3d_scatter(None, X1.tolist(), X2.tolist(), Y.tolist())
draw_plane(plt, regressor.coefficients, regressor.intercept, range=[(min(X1), max(X1)), (min(X2), max(X2))], elev=17, azim=70)
```

```python
regressor.coefficients, regressor.intercept
```

```python
draw_line(slope=float(regressor.coefficients[1]), intercept=float(regressor.intercept))
```

```python
sequential(None, [
  ['draw_scatter', {'x':X2.tolist(), 'y':Y.tolist()}],
  ['draw_line', {'slope':float(regressor.coefficients[1]), 'intercept':float(regressor.intercept), 'range': [0, max(X2.tolist() + Y.tolist())]}],
])
```
