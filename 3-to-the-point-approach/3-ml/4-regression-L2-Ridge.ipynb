```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

## L2 Ridge

Regularization Techinque to avoid overfitting

```python
dataset = pd.read_csv('../_/dataset/0-attrition.csv').select_dtypes(include='number')
```

```python
print('\n'.join(dataset.columns))
```

```python
input_cols = list(dataset.columns)
output_col = 'Monthly Income'
input_cols.remove('Employee ID')
input_cols.remove(output_col)
```

```python
x, y = np.array(dataset[input_cols]), np.array(dataset[output_col])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
plt.plot(x_train[:10])
```

```python
plt.plot(y_train[:10])
```

## Linear Model

```python
from sklearn.linear_model import LinearRegression

m = LinearRegression()
m.fit(x_train, y_train)
  
print(m.coef_)
plt.bar(np.linspace(1, len(m.coef_), len(m.coef_)), m.coef_)
```

```python
# skipping largers
c = []
for i in m.coef_:
  if i < 5: c.append(i)
  
print(c)
plt.bar(np.linspace(1, len(c), len(c)), c)
```

## L2 Model
Aka Ridge

```python
from sklearn.linear_model import Ridge

m = Ridge()
m.fit(x_train, y_train)
print(m.coef_)
plt.bar(np.linspace(1, len(m.coef_), len(m.coef_)), m.coef_)
```

```python
# skipping largers
c = []
for i in m.coef_:
  if i < 5: c.append(i)
  
print(c)
plt.bar(np.linspace(1, len(c), len(c)), c)
```
