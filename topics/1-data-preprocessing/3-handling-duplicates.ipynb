## Steps to Identify and Remove Duplicates
1. Identify Duplicates
1. Remove Duplicates
1. Handle Potential Issues

```python
import os
import pandas as pd
```

```python
# dataset
hotel_bookings = pd.read_csv(
  os.path.join(
    os.getcwd(),
    'datasets',
    'hotel-booking-demand/hotel_bookings.csv'
  )
)

hotel_bookings.describe()
```

```python
hotel_bookings.head()
```

```python
s = hotel_bookings.isna().sum()
s[s > 0]
```

```python
hotel_bookings.drop(columns=['company'], inplace=True) # too many missings
```

```python
s = hotel_bookings.isna().sum()
s[s > 0]
```

## 1. Identify Duplicates

```python
duplicates = hotel_bookings.duplicated()
duplicates_count = duplicates.sum()

print(duplicates_count)
duplicates_count / hotel_bookings.shape[0] * 100 # duplicate %age
```

## 2. Drop Duplicates

```python
hotel_bookings_cleaned = hotel_bookings.drop_duplicates()
```

```python
hotel_bookings_cleaned.describe()
```

```python
duplicates = hotel_bookings_cleaned.duplicated()
duplicates_count = duplicates.sum()

print(duplicates_count)
duplicates_count / hotel_bookings_cleaned.shape[0] * 100 # duplicate %age
```

## 3. Reset Index
you get index problem after removing duplicates

```python
print(
  'last_index =', hotel_bookings_cleaned.index[-1],
  'length =', hotel_bookings_cleaned.shape[0]
)
```

```python
hotel_bookings_cleaned.reset_index(drop=True, inplace=True)
```

```python
print(
  'last_index =', hotel_bookings_cleaned.index[-1],
  'length =', hotel_bookings_cleaned.shape[0]
)
```
